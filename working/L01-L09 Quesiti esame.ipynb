# L02_Q3

```python
import numpy as np
import time

def custom_std(a):
    mean = np.sum(a) / len(a)
    squared_diffs = np.sum((a - mean) ** 2)
    variance = squared_diffs / len(a)
    return np.sqrt(variance)

np.random.seed(0)
random_array = np.random.rand(1000000)

start_time = time.time()
custom_result = custom_std(random_array)
custom_duration = time.time() - start_time

start_time = time.time()
numpy_result = np.std(random_array)
numpy_duration = time.time() - start_time

print(f"Custom standard deviation: {custom_result}")
print(f"NumPy standard deviation: {numpy_result}")
print(f"Custom function runtime: {custom_duration:.6f} seconds")
print(f"NumPy function runtime: {numpy_duration:.6f} seconds")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def draw_circle(x0, y0, R, color):
    """
    Parameters:
        x0, y0: float - Center of the circle.
        R: float - Radius of the circle.
        color: str - Color of the circle.
    """
    theta = np.linspace(0, 2 * np.pi, 1000)  
    x = x0 + R * np.cos(theta) 
    y = y0 + R * np.sin(theta)  
    plt.fill(x, y, color=color, alpha=0.6)  

# Randomly draw 10 circles
plt.figure(figsize=(8, 8))
plt.axis("equal")

for _ in range(10):
    x0 = random.uniform(-10, 10)
    y0 = random.uniform(-10, 10)
    R = random.uniform(0.5, 3)
    color = np.random.rand(3,)
    
    draw_circle(x0, y0, R, color)

plt.title("Random Circles")
plt.show()
```

# L04_Q1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

def f(x):
    return np.exp(-x**2)

a, b = -5, 5
quad_result, quad_error = integrate.quad(f, a, b)
print(f"Reference integral (quad): {quad_result:.12f} with estimated error {quad_error:.2e}")

N_values = [32, 64, 128, 256, 512, 1024]
errors = []

for N in N_values:
    x = np.linspace(a, b, N)
    y = f(x)

    simpson_integral = integrate.simpson(y=y, x=x)
    
    error = abs(simpson_integral - quad_result)
    errors.append(error)
    
    print(f"N = {N:4d}, Simpson's Integral = {simpson_integral:.12f}, Error = {error:.2e}")

plt.figure(figsize=(8, 6))
plt.loglog(N_values, errors, marker='o', linestyle='-')
plt.title("Error of Simpson's Method vs Number of Sample Points (N)")
plt.xlabel("Number of Sample Points (N)")
plt.ylabel("Absolute Error")
plt.grid(True, which="both", ls="--")
plt.show()

print("\nReduction in error when N doubles:")
for i in range(1, len(N_values)):
    reduction = errors[i-1] / errors[i]
    print(f"N = {N_values[i-1]} to N = {N_values[i]}, Reduction Factor = {reduction:.2e}")
```

# L04_Q2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d, CubicSpline

def true_function(x):
    return np.sin(3 * x) + 0.5 * np.cos(5 * x) + 0.1 * x**2

N = 20
x_samples = np.linspace(-2, 2, N)
# x_samples = np.sort(np.random.uniform(-2, 2, N))  # Uncomment for irregular spacing
y_samples = true_function(x_samples)

x_test = np.linspace(-2, 2, 500)
y_true = true_function(x_test)

interp_methods = ['linear', 'quadratic', 'cubic']
interp_results = {}
for method in interp_methods:
    interpolator = interp1d(x_samples, y_samples, kind=method)
    interp_results[method] = interpolator(x_test)

cs_interpolator = CubicSpline(x_samples, y_samples)
cs_result = cs_interpolator(x_test)

errors = {}
for method in interp_methods:
    errors[method] = np.abs(interp_results[method] - y_true).max()
cs_error = np.abs(cs_result - y_true).max()

print("Interpolation Errors:")
for method in interp_methods:
    print(f"  {method.capitalize()}: {errors[method]:.6f}")
print(f"  Cubic Spline: {cs_error:.6f}")

plt.figure(figsize=(10, 6))
plt.plot(x_test, y_true, label="True Function", color='black', linewidth=2)
plt.scatter(x_samples, y_samples, color='red', label="Sampled Points", zorder=5)

for method in interp_methods:
    plt.plot(x_test, interp_results[method], label=f"Interp1d ({method})", linestyle='--')

plt.plot(x_test, cs_result, label="Cubic Spline", linestyle='-.')
plt.title("Interpolation Comparison")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

# L04_Q3

```python
import numpy as np
from scipy import optimize

def f(x):
    return x**3 - 2*x**2 - 11*x + 12

intervals = [(0, 2), (3, 5), (-5, -3)]
brentq_roots = []
for a, b in intervals:
    root = optimize.brentq(f, a, b)
    brentq_roots.append(root)

initial_guesses = [0, -10, 10]
root_results = []
for guess in initial_guesses:
    result = optimize.root(f, x0=guess)
    root_results.append(result.x[0])

brentq_verification = [f(root) for root in brentq_roots]
root_method_verification = [f(root) for root in root_results]

print("Roots found using Brent's method:")
for i, root in enumerate(brentq_roots):
    print(f"  Root {i+1}: {root:.6f}, f(root) = {brentq_verification[i]:.2e}")

print("\nRoots found using optimize.root:")
for i, root in enumerate(root_results):
    print(f"  Root {i+1}: {root:.6f}, f(root) = {root_method_verification[i]:.2e}")
```

# L07_Q1

```python
import numpy as np
import random
import time
from numba import njit

# States
states = ["Bull", "Bear", "Stagnant"]

# Transition matrix
P = np.array([
    [0.9, 0.075, 0.025],   # Bull market
    [0.15, 0.8, 0.05],     # Bear market
    [0.25, 0.25, 0.5]      # Stagnant market
])

# -------------------------------
# Pure Python version
# -------------------------------
def simulate_python(P, n_steps=10_000_000):
    n_states = len(P)
    current = random.randint(0, n_states - 1)
    counts = [0] * n_states

    for _ in range(n_steps):
        counts[current] += 1
        r = random.random()
        cum = 0.0
        for next_state, p in enumerate(P[current]):
            cum += p
            if r < cum:
                current = next_state
                break
    return np.array(counts) / n_steps


# -------------------------------
# Numba version
# -------------------------------
@njit
def simulate_numba(P, n_steps=10_000_000):
    n_states = P.shape[0]
    current = np.random.randint(0, n_states)
    counts = np.zeros(n_states, dtype=np.int64)

    for _ in range(n_steps):
        counts[current] += 1
        r = np.random.random()
        cum = 0.0
        for next_state in range(n_states):
            cum += P[current, next_state]
            if r < cum:
                current = next_state
                break
    return counts / n_steps


# -------------------------------
# Run and compare
# -------------------------------
if __name__ == "__main__":
    n_steps = 5_000_000

    t0 = time.time()
    pi_py = simulate_python(P, n_steps)
    t1 = time.time()

    t2 = time.time()
    pi_nb = simulate_numba(P, n_steps)
    t3 = time.time()

    print("Python result:", pi_py)
    print("Numba  result:", pi_nb)
    print(f"Python time: {t1 - t0:.2f}s")
    print(f"Numba  time: {t3 - t2:.2f}s")
```

```python
import matplotlib.pyplot as plt
from functools import wraps

def myplot(func):
    """Decorator that standardizes matplotlib style and saves output as PDF."""
    @wraps(func)
    def wrapper(*args, **kwargs):
        plt.rcParams.update({
            "figure.figsize": (6, 4),
            "font.size": 14,
            "axes.labelsize": 14,
            "axes.titlesize": 16,
            "xtick.labelsize": 12,
            "ytick.labelsize": 12,
            "legend.fontsize": 12,
            "lines.linewidth": 2,
            "axes.grid": True,
            "grid.alpha": 0.3,
        })

        fig, ax = plt.subplots()
        result = func(fig, ax, *args, **kwargs)

        filename = f"{func.__name__}.pdf"
        fig.tight_layout()
        fig.savefig(filename, bbox_inches="tight")
        print(f"✅ Figure saved as {filename}")

        return fig 
    return wrapper

@myplot
def plot_spectrum(fig, ax, data=None):
    import numpy as np
    if data is None:
        x = np.linspace(0, 10, 200)
        y = np.sin(x) * np.exp(-0.1 * x)
    else:
        x, y = data
    ax.plot(x, y, label="Signal")
    ax.set_xlabel("Energy [MeV]")
    ax.set_ylabel("Counts")
    ax.legend()
    ax.set_title("Energy Spectrum")
    #return fig

# Run it
plot_spectrum()

```

# L09_Q1

```python
import sympy as sym
from sympy import init_session
init_session()

# Initialize symbolic variable
x = sym.Symbol('x')

# Define the expression: f = x*e^(-x) + x*(1 - x)
f = x * sym.exp(-x) + x * (1 - x)
print("Expression f(x):")
sym.pprint(f)

# Evaluate at specific values of x
values = [0, 0.1, 0.2, 0.4, 0.8]
print("\nEvaluations:")
for val in values:
    numeric_val = f.subs(x, val).evalf()  # substitute x = val and evaluate numerically
    print(f"f({val}) = {numeric_val:.6f}")
```

```python
import sympy as sym
from sympy import init_session
init_session()

# Define polynomial
poly = x**4 - 6*x**3 + x**2 + 24*x + 16
print("\nPolynomial:")
sym.pprint(poly)

# Factor the polynomial
factored_poly = sym.factor(poly)
print("\nFactored form:")
sym.pprint(factored_poly)

# Find its zeros
roots = sym.solve(poly, x)
print("\nZeros of the polynomial:")
for r in roots:
    print(r.evalf())
```

```python
import sympy as sym
from sympy import init_session
init_session()

# Define the function
f = sym.sin(x) * sym.exp(-x)
print("\nFunction f(x):")
sym.pprint(f)

# Integrate the function
F = sym.integrate(f, x)
print("\nIntegral F(x):\n\n")
sym.pprint(F.simplify())

# Differentiate back
dF = sym.diff(F, x)
print("\n\n\nDerivative of F(x):\n\n")
sym.pprint(dF.simplify())

# Check equivalence
print("\nCheck if derivative equals original:")
print(sym.simplify(dF - f) == 0)
```
